```python
import hashlib, os

nombre_archivo = "00_setup_PabloMonserratSastre.ipynb"

if not os.path.exists(nombre_archivo):
    with open(nombre_archivo, "w") as f:
        f.write("Plantilla reproducible - sustituir por el .ipynb real si se descarga")
    print("Archivo provisional creado (solo para cálculo de hash).")

with open(nombre_archivo, "rb") as f:
    contenido = f.read()

sha = hashlib.sha256(contenido).hexdigest()
print("SHA-256 generado:\n", sha)
```

```python
import json, datetime

registro = {
    "archivo": nombre_archivo,
    "sha256": sha,
    "fecha": datetime.datetime.now().isoformat(),
    "autor": "Pablo Monserrat Sastre"
}

with open("registro_hash.json", "w", encoding="utf-8") as f:
    json.dump(registro, f, indent=2, ensure_ascii=False)

print("Registro creado: registro_hash.json")
```

```python
with open(nombre_archivo, "rb") as f:
    nuevo_hash = hashlib.sha256(f.read()).hexdigest()

print("Hash original:", sha)
print("Hash recalculado:", nuevo_hash)
print("Coinciden" if nuevo_hash == sha else "Error – archivo modificado")
```

```python
EEE = {
    "Exactitud": 2,
    "Explicabilidad": 2,
    "Economía": 2,
    "Comentario": "Notebook reproducible con hash verificado y documentación completa"
}
print("Autoevaluación EEE:", EEE)
```
